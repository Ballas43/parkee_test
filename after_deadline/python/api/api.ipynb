```python
import subprocess
import pandas as pd
import json
import numpy as np
```

```python
cmd =  ["curl", "http://universities.hipolabs.com/search?country=United%20States"]

process = subprocess.run(cmd, stdout=subprocess.PIPE, stderr=subprocess.DEVNULL)

data = process.stdout.decode()
```

```python
data_js = json.loads(data)
data_js
```

```python
data_pd = pd.DataFrame(data_js)
data_pd
```

```python
data_pd.rename(columns={'name':'Name', 'web_pages':'Web pages', 'country':'Country', 'domains':'Domains', 'state-province':'State Province'}, inplace=True)
data_pd
```

```python
data_pd.drop(columns='alpha_two_code', inplace=True)
data_pd
```

```python
data_pd_ordered = data_pd[['Name', 'Web pages', 'Country', 'Domains', 'State Province']]
data_pd_ordered
```

```python
data_pd_ordered = data_pd_ordered.replace(to_replace='None', value=np.nan).dropna()
data_pd_ordered
```
