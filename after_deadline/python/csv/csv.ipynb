```python
import pandas as pd
import numpy as np
```

```python
df = pd.concat(map (pd.read_csv, ['branch_a.csv', 'branch_b.csv', 'branch_c.csv']), ignore_index=True)

df
```

```python
df = df.dropna()
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df
```

```python
df = df.drop_duplicates(subset=['transaction_id'])
df
```

```python
df_sales = df.copy(deep=True)
df_sales
```

```python
df_sales["sales"] = df_sales['quantity'] * df_sales['price']
df_sales
```

```python
branch = df_sales['branch'].unique()
sales = []

for data in branch:
    sales.append(df_sales.loc[df_sales['branch'] == data, 'sales'].sum())

sales, branch
```

```python
new_data = np.stack((branch, sales), axis=1)
new_data
```

```python
sales_df = pd.DataFrame(new_data, columns=['Branch', 'Total'])
sales_df
```

```python
sales_df.to_csv('total_sales_per_branch.csv', index=False)
```
